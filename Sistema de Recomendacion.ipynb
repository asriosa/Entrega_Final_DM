# Parte Sistema de Recomendacion con Filtrado Colaborativo

```python
#Subset de registros Distritos
data_seleccionada.shape
```

```python
#Districo seleccionado como perfil
registro.shape
```

```python
# Quitamos en ambos las caolumnas no numericas, ya que ára hacer la comparacion entre mi usuario consulta 
#(distrito y el sub sed, los datos ddeben estan en valor numerico)
data_seleccionada = data_seleccionada.select_dtypes(exclude=['object'])
data_seleccionada.fillna(0, inplace=True)

registro = registro.select_dtypes(exclude=['object'])
registro.fillna(0, inplace=True)
```

```python
data_seleccionada.shape
```

```python
registro.shape
```

```python
registro.head(1)
```

```python
df_distritos = data_seleccionada.drop(['Unnamed: 0', 'YEAR','UBIGEO','latitud','longitud','cluster_std'], axis=1)
df_distritos
```

```python
df_un_distrito = registro.drop(['Unnamed: 0', 'YEAR','UBIGEO','latitud','longitud','cluster_std'], axis=1)
df_un_distrito
```

## Similaridad de Coseno

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calcular la similitud del coseno entre el distrito seleccionado y cada registro en el subconjunto
cosine_similarities = cosine_similarity(df_un_distrito, df_distritos)

# Crear un DataFrame con los resultados y asignar el resultado a una columna en df_distritos
df_distritos['Similitud_coseno'] = cosine_similarities.flatten()
```

```python
df_distritos.head(2)
```

```python
df_distritos_sorted = df_distritos.sort_values(by='Similitud_coseno', ascending=False)
```

```python
df_distritos_sorted.columns
```

```python
df_distritos_sorted.head(4)
```

## Similaridad de Person

```python
# Excluir la columna 'Similitud_coseno' de las columnas numéricas
columnas_numericas = df_distritos.select_dtypes(include=['float64', 'int64']).columns
columnas_numericas = columnas_numericas.difference(['Similitud_coseno'])

# Calcula la correlación de Pearson para el registro seleccionado con respecto a df_distritos
correlaciones = df_distritos[columnas_numericas].apply(lambda row: row.corr(df_un_distrito.iloc[0][columnas_numericas]), axis=1)

# Agrega la columna de correlaciones al DataFrame del subset
df_distritos['Correlacion_Pearson'] = correlaciones

# Ordena el DataFrame de subset de mayor a menor basado en la correlación de Pearson
df_distritos_corre = df_distritos.sort_values(by='Correlacion_Pearson', ascending=False).copy()
```
