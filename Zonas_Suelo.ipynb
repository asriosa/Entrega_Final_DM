**Mapa de la distribución del peligro sísmico en el Perú.**

Diccionario:

Ubigeo: Código de Ubicación Geográfica.

Zona: Zona sísmica de cada ubigeo que indica su vulnerabilidad.

*   Zona 4: Zona con mayor vulnerabilidad ante actividad sísmica.
*   Zona 3: Zona con una vulnerabilidad considerable ante actividad sísmica.
*   Zona 2: Zona con una vulnerabilidad moderada ante actividad sísmica.
*   Zona 1: Zona con menor vulnerabilidad ante actividad sísmica.









```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('suelo_nivel.csv')
df
```

```python
df.ZONA.value_counts()
```

```python
# Este código crea un gráfico de pastel que muestra la distribución de categorías I1 a I4 de establecimientos de salud en Perú para el 2022.
# Calcula la suma de las categorías I1 hasta I4 en todos los distritos
total_1 = df.ZONA[df['ZONA'] == 1].count()
total_2 = df.ZONA[df['ZONA'] == 2].count()
total_3 = df.ZONA[df['ZONA'] == 3].count()
total_4 = df.ZONA[df['ZONA'] == 4].count()

explode_list = [0, 0, 0.05, 0] # ratios para alejar las porciones (continentes) del centro.
# Crea un gráfico de pastel
plt.figure(figsize=(10, 8))
plt.pie([total_1, total_2, total_3, total_4],
        labels=['1','2','3','4'], shadow=True,
        autopct='%1.1f%%', startangle=90,explode=explode_list, pctdistance=0.5)
plt.title("Distribución de zonas 1 a 4 de tipos de suelo en el Perú")
plt.axis('equal')  # Para que el pie no se sobreponga con la leyenda.
plt.legend(title = 'Categorías',labels = ['1', '2', '3', '4'], loc='upper left')  # añade leyenda

# El gráfico muestra que la categoría I1 es la más abundante, seguida de las categorías I2 y I4. La categoría I3 es la menos abundante.
plt.show()
```

El gráfico permite identificar que en el Perú predomina el suelo con un nivel 3 de riesgo eon 42.4%, es decir, las zonas con una vulnerabilidad considerable ante un sismo.

```python
ubigeos = pd.read_csv("TB_UBIGEOS.csv")
ubigeos
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import geopy.distance
```

```python
# Unir las tablas
zonas = pd.merge(df, ubigeos, right_on='ubigeo_inei', left_on = 'ubigeo', how='left')

# Obtener las columnas de latitud y longitud de la segunda tabla
zonas = zonas[['nombdep', 'nombprov', 'nombdist', 'ubigeo', 'ZONA','latitud', 'longitud']]

# Imprimir la tabla unida
zonas
```

```python
# Crear un GeoDataFrame a partir de los datos de latitud, longitud y cluster_std
gdf = gpd.GeoDataFrame(zonas, geometry=gpd.points_from_xy(zonas['longitud'], zonas['latitud']))

# Cargar los límites de Perú
peru_boundaries = gpd.read_file('peru_shapes/per_admbnda_adm1_ign_20200714.shp')

colors = {1: 'blue', 2: 'green', 3: 'yellow', 4:'red'}  # Ajusta los colores según tus preferencias

# Crear un mapa de dispersión de los datos geoespaciales con colores codificados por cluster_std
plt.figure(figsize=(10, 10))

for cluster_value, color in colors.items():
    subset = gdf[gdf['ZONA'] == cluster_value]
    plt.scatter(subset.geometry.x, subset.geometry.y, color=color, label=f'Zona {cluster_value}', marker='o', edgecolors='black', linewidths=0.5)

# Agregar límites de distritos
peru_boundaries.boundary.plot(ax=plt.gca(), linewidth=1, color='black')

plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.title('Mapa de la distribución del peligro sísmico en el Perú')
plt.axis('off')
plt.legend(loc='upper left')
plt.show()
```
